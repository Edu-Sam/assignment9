```python
heart_continous_array=heart["oldpeak"].to_numpy()
heart_continous_array
```

```python
mean = np.mean(heart_continous_array) 
std = np.std(heart_continous_array) 
print('mean of the dataset is', mean) 
print('std. deviation is', std) 

threshold = 3
outlier = [] 
for i in heart_continous_array: 
    z = (i-mean)/std 
    if z > threshold: 
        outlier.append(i) 
print('outlier in dataset is', outlier) 
```
